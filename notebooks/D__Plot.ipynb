# Plotting a Status Layer in Matplotlib

```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show
from dist_s1.dist_plot import add_dist_s1_colorbar, get_dist_s1_mpl_cmap
```

```python
with rasterio.open('plot_example/OPERA_L3_DIST-ALERT-S1_T19HBD_20240402T100430Z_20251114T014222Z_S1A_30_v0.1_GEN-DIST-STATUS.tif') as ds:
    X = ds.read(1)
    p = ds.profile
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 10), gridspec_kw={'width_ratios': [.95, 0.05]})

cmap, norm = get_dist_s1_mpl_cmap()
show(X, ax=axs[0], transform=p['transform'], cmap=cmap, norm=norm)
add_dist_s1_colorbar(axs[1])
plt.show()
```
